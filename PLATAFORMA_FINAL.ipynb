```python
# ── 2. Escribir el código de la app ──────────────────────────────────────────
import base64, os

APP_CODE_B64 = """"""

os.makedirs("/content/app", exist_ok=True)
with open("/content/app/app.py", "wb") as f:
    f.write(base64.b64decode(APP_CODE_B64))
print("✅ app.py listo — navegación top-bar, sin sidebar")
```

```python
# ── 3. Subir archivos de datos ────────────────────────────────────────────────
from google.colab import files
import shutil, os, re

os.makedirs("/content/app", exist_ok=True)

# Copiar si ya existen en /content/
for fname in ["EXPORTACIONES.xlsx", "IMPORTACIONES.xlsx", "PANEL.csv"]:
    src = f"/content/{fname}"
    dst = f"/content/app/{fname}"
    if os.path.exists(src) and not os.path.exists(dst):
        shutil.copy(src, dst)
        print(f"  📋  Copiado desde /content/: {fname}")

required = ["EXPORTACIONES.xlsx", "IMPORTACIONES.xlsx", "PANEL.csv"]
missing  = [r for r in required if not os.path.exists(f"/content/app/{r}")]

if missing:
    print(f"⬆️  Faltan archivos: {missing} — súbelos aquí:")
    uploaded = files.upload()
    for fname, data in uploaded.items():
        clean = re.sub(r'\s*\(\d+\)', '', fname).strip()
        dest = f"/content/app/{clean}"
        with open(dest, "wb") as f:
            f.write(data)
        print(f"  ✅  {fname} → {dest}")

missing2 = [r for r in required if not os.path.exists(f"/content/app/{r}")]
if missing2:
    print(f"\n⚠️  Aún faltan: {missing2}")
else:
    print("\n✅ Todos los archivos listos. Continúa con la Celda 4.")
```

```python
# ── 4. Lanzar la app Streamlit con túnel público ─────────────────────────────
import subprocess, time, os
from pyngrok import ngrok

os.system("pkill -f streamlit 2>/dev/null; sleep 1")
ngrok.kill()

PORT = 8501
env = os.environ.copy()
env["APP_DATA_DIR"] = "/content/app"

proc = subprocess.Popen(
    ["streamlit", "run", "/content/app/app.py",
     "--server.port", str(PORT),
     "--server.headless", "true",
     "--server.enableCORS", "false",
     "--server.enableXsrfProtection", "false",
     "--browser.gatherUsageStats", "false"],
    stdout=subprocess.DEVNULL,
    stderr=subprocess.DEVNULL,
    env=env,
)

print("⏳ Iniciando servidor Streamlit…")
time.sleep(8)

ngrok.set_auth_token("3AVECrkRTZAy5EstrFgdmGN5lYI_CTcXtZ2wjUNu5Mf2sEFi")
tunnel = ngrok.connect(PORT)
public_url = tunnel.public_url

print()
print("=" * 60)
print("  🚀  APP LISTA")
print(f"  🔗  {public_url}")
print("  Abre el link en una nueva pestaña.")
print("=" * 60)
```
